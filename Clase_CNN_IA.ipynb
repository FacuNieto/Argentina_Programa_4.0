## 1.3 Filtrado

```python
# Cargamos una imagen
edificio = cv2.imread('CNN_IA_FCEFyN/images/edificio.jpg')
edificio = cv2.cvtColor(edificio, cv2.COLOR_BGR2GRAY)
plt.imshow(edificio)
plt.colorbar()
plt.show()
```

```python
# Definimos algunos filtros:
kernel3x3_average = np.array([[1/9, 1/9, 1/9],
                              [1/9, 1/9, 1/9],
                              [1/9, 1/9, 1/9]])

kernel3x3_horiz = np.array([[-1, -1, -1],
                            [ 2,  2,  2],
                            [-1, -1, -1]])

kernel3x3_vertic = np.array([[-1,  2, -1],
                             [-1,  2, -1],
                             [-1,  2, -1]])

kernel3x3_45 = np.array([[-1, -1,  2],
                         [-1,  2, -1],
                         [ 2, -1, -1]])

kernel3x3_lap = np.array([[ 0, -1,  0],
                          [-1,  4, -1],
                          [ 0, -1,  0]])

# https://aishack.in/tutorials/image-convolution-examples/ varios tipos de filtros!
```

```python
# Visualizamos los filtros!
plt.figure(figsize=(25,5))
plt.subplot(141)
plt.title('kernel3x3_horiz')
plt.imshow(kernel3x3_horiz)
plt.colorbar()
plt.subplot(142)
plt.title('kernel3x3_vertic')
plt.imshow(kernel3x3_vertic)
plt.colorbar()
plt.subplot(143)
plt.title('kernel3x3_45')
plt.imshow(kernel3x3_45)
plt.colorbar()
plt.subplot(144)
plt.title('kernel3x3_lap')
plt.imshow(kernel3x3_lap)
plt.colorbar()
plt.show()
```

```python
# Aplicamos los filtros a la imagen
difuminado = cv2.filter2D(src=edificio, ddepth=-1, kernel=kernel3x3_average)
horiz=cv2.filter2D(src=edificio, ddepth=-1, kernel=kernel3x3_horiz)
vertic=cv2.filter2D(src=edificio, ddepth=-1, kernel=kernel3x3_vertic)
im_45=cv2.filter2D(src=edificio, ddepth=-1, kernel=kernel3x3_45)
lap=cv2.filter2D(src=edificio, ddepth=-1, kernel=kernel3x3_lap)
```

```python
# Visualizamos los resultados del filtrado
plt.figure(figsize=(20,10))
plt.subplot(231)##########
plt.title('Sin Filtro')
plt.imshow(edificio)
plt.subplot(232)##########
plt.title('Difuminado')
plt.imshow(difuminado)
plt.subplot(233)##########
plt.title('Horiz')
plt.imshow(horiz)
plt.subplot(234)##########
plt.title('Vertical')
plt.imshow(vertic)
plt.subplot(235)##########
plt.title('im_45')
plt.imshow(im_45,cmap='Blues')
plt.subplot(236)##########
plt.title('Filtro operador laplaciano')
plt.imshow(lap)
plt.show()
```

## 2.1 Cargar Dataset

```python
#importar librerias
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import numpy as np
import matplotlib.pyplot as plt
```

```python
#dividimos en train y test
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

## 2.3 Escalar, reshape y encoding

```python
train_images = train_images.astype('float32') / 255
test_images = test_images.astype('float32') / 255
```

```python
train_images = train_images.reshape(train_images.shape[0],28, 28, 1) # el 1 es de un canal ya que estamos en escalda de grises, si fuera RGB seria 3.
test_images = test_images.reshape(test_images.shape[0], 28, 28, 1)
```

```python
train_labels = tf.keras.utils.to_categorical(train_labels, 10) # One hot encoding 
test_labels = tf.keras.utils.to_categorical(test_labels, 10)
```

```python
train_images.shape
```

```python
train_labels.shape
```

```python
print(train_labels[0])
```

## 2.4 Definimos arquitectura, compilamos y entrenamos

```python
model = tf.keras.models.Sequential([
    
    # 1°Conv, definimos entrada de 28,28 escala de grises
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D(2, 2),

    # 2°Conv
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    
    # Flatten
    tf.keras.layers.Flatten(),
    # Densa de 128 neuronas 
    tf.keras.layers.Dense(128, activation='relu'),
    # 10 Clases
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy', 
              optimizer='adam', 
              metrics=['accuracy'])
```

Losses:

BinaryCrossentropy: Computes the cross-entropy loss between true labels and predicted labels.

BinaryFocalCrossentropy: Computes the focal cross-entropy loss between true labels and predictions.

CategoricalCrossentropy: Computes the crossentropy loss between the labels and predictions.

CategoricalHinge: Computes the categorical hinge loss between y_true and y_pred.

CosineSimilarity: Computes the cosine similarity between labels and predictions.

Hinge: Computes the hinge loss between y_true and y_pred.

Huber: Computes the Huber loss between y_true and y_pred.

KLDivergence: Computes Kullback-Leibler divergence loss between y_true and y_pred.

LogCosh: Computes the logarithm of the hyperbolic cosine of the prediction error.

Loss: Loss base class.

MeanAbsoluteError: Computes the mean of absolute difference between labels and predictions.

MeanAbsolutePercentageError: Computes the mean absolute percentage error between y_true and y_pred.

MeanSquaredError: Computes the mean of squares of errors between labels and predictions.

MeanSquaredLogarithmicError: Computes the mean squared logarithmic error between y_true and y_pred.

Poisson: Computes the Poisson loss between y_true and y_pred.

Reduction: Types of loss reduction.

SparseCategoricalCrossentropy: Computes the crossentropy loss between the labels and predictions.

SquaredHinge: Computes the squared hinge loss between y_true and y_pred.

https://www.tensorflow.org/api_docs/python/tf/keras/losses

```python
history = model.fit(train_images, train_labels,epochs = 10, 
                    validation_data = (test_images, test_labels), 
                    verbose = 1, 
                    )
```

### 2.4.1 Curvas de entrenamiento

```python
# Función para graficar
def plot_graphs(history, string):
  plt.plot(history.history[string], color='blue')
  plt.plot(history.history['val_'+string], color='red')
  plt.xlabel("Epochs")
  plt.ylabel(string)
  plt.legend([string, 'val_'+string])
  plt.show()

# Graficar curvas de loss y accuracy para datos de entrenamiento y validación
plot_graphs(history, 'accuracy')
plot_graphs(history, 'loss')
```

### 2.4.2 Evaluamos

```python
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print('\nTest accuracy:', test_acc)
```

## 2.6 Deconstruimos la red

### Filtros, kernels? 

```python
# obtener datos de las layers
for layer in model.layers:
	# print(layer.name)
	# check for convolutional layer
	if 'conv' not in layer.name:
		continue
	# get filter weights
	filters, biases = layer.get_weights()
	print(layer.name, filters.shape)
```

```python
filters, biases = model.layers[0].get_weights() # PRIMER CAPA! 

# normalize filter values to 0-1 so we can visualize them
f_min, f_max = filters.min(), filters.max()
filters = (filters - f_min) / (f_max - f_min)
```

```python
filters.shape
```

```python
# plot first few filters
n_filters, ix = 1,1
for i in range(n_filters):
	# get the filter
	f = filters[:, :, :, i]
	# plot each channel separately
	for j in range(1):
		# specify subplot and turn of axis
		ax = plt.subplot(n_filters, 2, ix)
		ax.set_xticks([])
		ax.set_yticks([])
		# plot filter channel in grayscale
		plt.imshow(f[:, :, j])
		ix += 1
# show the figure
plt.colorbar()
plt.show()

ker=pd.DataFrame(np.array(f[:, :, 0]))
display(ker)
# fuente: https://machinelearningmastery.com/how-to-visualize-filters-and-feature-maps-in-convolutional-neural-networks/ 
```
